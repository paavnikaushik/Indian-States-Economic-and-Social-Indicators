```python
#importing required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
#reading the csv file
df=pd.read_csv(r"C:\Users\paavn\Desktop\data_india.csv")
```

```python
#according to the statistics, the average household size across states is between 5-6 members,around 87% of the people own the houses they live in and around 43% of the people have received no schooling.
#while, agriculture still remains the largest employer, having 46% of the people on average engaged in agriculture
# the average income is near 10K per year in india. 
# Around 60% of the people consider themselves physically active.
# In diseases, hypertension is seen to affect around 30% of the population which is a significant percentage of the whole population.
# A high disparity can be see within the states as the range is significantly large for almost all the indicators across social and health domains
```

```python
#resetting index as state name
df.set_index('Indicators',inplace=True)
```

```python
#sort values by GDP
sort_df=df.sort_values(by='GDP IN LACS',ascending=False)
sort_df
```

```python
#plotting the barplot for the top 15 states by GDP
plt.figure(figsize=(10, 6))
sns.barplot(x='Indicators',y='GDP IN LACS',data=sort_df,color='red',palette='rocket',hue='Indicators')
plt.xlabel('State')
plt.ylabel('GDP')
plt.title('Top 15 States by GDP')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
#Maharashtra,Gujarat,Tamil Nadu are the top 3 states in terms of GDP measured at constant prices
```
