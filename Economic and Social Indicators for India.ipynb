```python
#sort values by GDP
sort_df=df.sort_values(by='GDP IN LACS',ascending=False)
sort_df
```

# Barplot showing top 15 states by GDP

```python
#plotting the barplot for the top 15 states by GDP
plt.figure(figsize=(10, 6))
sns.barplot(x='Indicators',y='GDP IN LACS',data=sort_df,color='red',palette='rocket',hue='Indicators',legend=False)
plt.xlabel('State')
plt.ylabel('GDP')
plt.title('Top 15 States by GDP')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.show()
```

```python
#Maharashtra,Gujarat,Tamil Nadu are the top 3 states in terms of GDP measured at constant prices
```

# Correlations and Heatmap



```python
#Calculating correlations
df=df.set_index('Indicators')
df.corr()
```

```python
#Presenting the results in the form of a heatmap
heatmap=sns.heatmap(df.corr(),cbar=True,linewidth=0.5,cmap='inferno',annot=False,linewidths=4)
heatmap.set_title("Correlation Heatmap between Indicators")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
#There is a high positive correlation between average people in the house and households owning the residence which might hint that the tendency to live in rented homes is seen more in small family sizeshowever more data is needed on small family sizes to arrive at a conclusion
#There is a high positive correlatio between people engaged in agriculture and percentage who received no schooling
# If we look at people engaged in agriculture, there is a negative correlation between the occurences of the disease and the percentage of people engaged in agriculture, that means that as the number of people engaged in agriculture increase, the number of occurences of these diseases go down
# There is particularly high positive correlation between the wage and salaried workers' percentage and the incidence of diseases. As the salaried working class is increasing in our country, it might explain the rise in the incidences of such diseases.
# There is a negative relationship between the percent who consider themselves physically active and the mean income from all sources, it might explain that the more laboursome tasks are given less pay across states.
```

2. Correlation Analysis Insights:

 Key Implications:
- Small family sizes may prefer rented homes.
- Link between agriculture and education levels.
- Health benefits associated with agricultural engagement.
- Caution needed with the rise in diseases among wage/salary workers.
- Physically demanding jobs may need better compensation.

Recommendations:
- Target health interventions based on occupation.
- Promote education for those engaged in agriculture.
- Address disparities in income for physically active occupations.
